### Q-6. A pharmaceutical company is developing a drug that is supposed to reduce blood pressure. They conduct a clinical trial with 100 patients and record their blood pressure before and after taking the drug. The company wants to know if the change in blood pressure follows a normal distribution.

```python
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data (1).csv")
df.head()
```

```python
df.rename(columns = {' Blood Pressure Before (mmHg)':'Blood Pressure Before (mmHg)', ' Blood Pressure After (mmHg)': 'Blood Pressure After (mmHg)'}, inplace = True)
```

```python
df["change in blood pressure"] = df["Blood Pressure Before (mmHg)"]-df["Blood Pressure After (mmHg)"]
```

```python
df.head()
```

```python
sns.histplot(df["change in blood pressure"], kde=True)
```

```python
variable = df['change in blood pressure']

# Null hypothesis: The variable is normally distributed
# Alternative hypothesis: The variable is not normally distributed

statistic, p_value = stats.shapiro(variable)

print(f"Test statistic: {statistic}")
print(f"P-value: {p_value}\n")

if p_value < 0.05:
    print("Reject the null hypothesis at the 95% confidence level.")
    print("Data is not normally distributed.")
else:
    print("Fail to reject the null hypothesis at the 95% confidence level.")
    print("Data is normally distributed.")
```

```python
stats.probplot(df["change in blood pressure"], plot=plt)
plt.show()
```
